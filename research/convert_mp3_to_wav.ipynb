```python
from os import path
from pydub import AudioSegment
import os
from tqdm.notebook import tqdm
```

```python
data = {}
audio_dir = '../tatar_tts/'
```

```python
for filename in os.listdir(audio_dir):
    name = os.path.splitext(filename)[0]
    if name not in data:
        data[name] = {
            'mp3': name + '.mp3',
            'wav': name + '.wav'
        }
```

```python
for name in tqdm(data):
    sound = AudioSegment.from_mp3(audio_dir + data[name]['mp3'])
    sound.export(audio_dir + data[name]['wav'], format="wav")
```
